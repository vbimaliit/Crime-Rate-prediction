What is the best alpha?

```python
model5.alpha_
```

Alpha, generally known regularization; is used to set certain penalty so that the noise in the data-set can be minimized. Large values for alpha can result in regularization effect which dominates the squared loss function and the coefficients curves to zero. If the alpha value curves to zero and the solution inclines towards the ordinary least squares then large oscillations can be observed in coefficients. Hence, alpha should be selected such that balance is maintained between coefficients and penalty.

What does this say about the amount of overfitting in linear regression for this problem?


MSE is the major factor used in regression model determining whether the data is under-fitting or over-fitting the 
model. In linear regression as the we move towards next data points due to lack of biasness the data over-fits the 
model. However in Ridge regression, which is a regularized form of linear regression model which sets alpha as a 
penalty value which creates biasness preventing the model from over-fitting.
